## Uber VS Sharenow

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.gridspec import GridSpec
import numpy as np
import pandas as pd

sharenow = pd.read_excel("sharenow.xlsx", header=5)
uber = pd.read_excel("uber.xlsx", header=5)
plt.rcParams["figure.figsize"] = (26,8)
plt.rcParams['savefig.dpi'] = 120
```

### Tasso di engagement

```python
x = range(1, 31)
y_sh = sharenow["Tasso di engagement"]
y_ub = uber["Tasso di engagement"]
fig, ax1 = plt.subplots()
ye_sh_a = np.array(y_sh)
ye_ub_a = np.array(y_ub)
max_ye_sh = max(ye_sh_a)
max_xe_sh = x[ye_sh_a.argmax()]
max_ye_ub = max(ye_ub_a)
max_xe_ub = x[ye_ub_a.argmax()]
ax1.set_title("Tasso di engagement")
ax1.set_xlabel("Giorni")
ax1.set_ylabel("Engagement")
ax1.tick_params(axis="both")
ax1.plot(x, y_sh, color='b', label='Sharenow', linewidth=1)
ax1.plot(x, y_ub, color='g', label='Uber', linewidth=1)
ax1.annotate('Massimo Engagement Sharenow', xy=(max_xe_sh, max_ye_sh),
            xytext=(0.8, 0.9), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05, width=1),
            horizontalalignment='right', verticalalignment='top',
            )
ax1.annotate('Massimo Engagement Uber', xy=(max_xe_ub, max_ye_ub),
            xytext=(0.35, 0.5), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05, width=1),
            horizontalalignment='right', verticalalignment='top',
            )
ax1.legend()
plt.show()
```

### Obiettivi dei post

```python
sharenow_purpose = sharenow.groupby("Obiettivo Post (ricorrenza, real time, vendita prodotto, ecc)").size()
labels_sh = list(*sharenow_purpose.axes)
labels_sh[3], labels_sh[4] = labels_sh[4], labels_sh[3]
sizes_sh = list(sharenow_purpose)
uber_purpose = uber.groupby("Obiettivo Post (ricorrenza, real time, vendita prodotto, ecc)").size()
labels_ub = list(*uber_purpose.axes)
sizes_ub = list(uber_purpose)
fig1, axs = plt.subplots(1, 2)
axs[0].set_title("Obiettivi post Sharenow")
axs[0].pie(sizes_sh, labels=None, autopct='%1.1f%%', startangle=90, wedgeprops={'linewidth': 1, 'edgecolor': 'white'})
axs[0].axis('equal')
axs[1].set_title("Obiettivi post Uber")
axs[1].pie(sizes_ub, labels=None, autopct='%1.1f%%', startangle=90, wedgeprops={'linewidth': 1, 'edgecolor': 'white'})
axs[1].axis('equal')
axs[0].legend( loc="best", labels=labels_sh) 
plt.show()
```

## Numero Commenti e Likes

```python
y1_sh = sharenow['Numero di "Mi Piace"']
y1_ub = uber['Numero di "Mi Piace"']
y2_sh = sharenow['Numero di commenti']
y2_ub = uber['Numero di commenti']
y1_sh_a = np.array(y1_sh)
y1_ub_a = np.array(y1_ub)
y2_sh_a = np.array(y2_sh)
y2_ub_a = np.array(y2_ub)
max_y_sh = max(y1_sh_a)
max_x_sh = x[y1_sh_a.argmax()]
max_y_ub = max(y1_ub_a)
max_x_ub = x[y1_ub_a.argmax()]
max_x1_sh = x[y2_sh_a.argmax()]
max_y1_sh = max(y2_sh_a)
max_x1_ub = x[y2_ub_a.argmax()]
max_y1_ub = max(y2_ub_a)
```

```python
fig2, axs1 = plt.subplots(1, 2)
axs1[0].set_title("Numero di Likes")
axs1[0].set_xlabel("Giorni")
axs1[0].set_ylabel("Likes")
axs1[0].tick_params(axis="both")
axs1[0].plot(x, y1_sh, color='b', label='Sharenow', linewidth=1)
axs1[0].plot(x, y1_ub, color='g', label='Uber', linewidth=1)
axs1[1].set_title("Numero di commenti")
axs1[1].set_xlabel("Giorni")
axs1[1].set_ylabel("Commenti")
axs1[1].plot(x, y2_sh, color='b', label='Sharenow', linewidth=1)
axs1[1].plot(x, y2_ub, color='g', label='Uber', linewidth=1)
axs1[0].annotate('Massimo dei "Mi Piace" Sharenow', xy=(max_x_sh, max_y_sh),
            xytext=(0.8, 0.3), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05, width=1),
            horizontalalignment='right', verticalalignment='top',
            )
axs1[0].annotate('Massimo dei "Mi Piace" Uber', xy=(max_x_ub, max_y_ub),
            xytext=(0.3, 0.9), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05, width=1),
            horizontalalignment='right', verticalalignment='top',
            )
axs1[1].annotate('Massimo dei Commenti Sharenow', xy=(max_x1_sh, max_y1_sh),
            xytext=(0.8, 0.15), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05, width=1),
            horizontalalignment='right', verticalalignment='top',
            )
axs1[1].annotate('Massimo dei Commenti Uber', xy=(max_x1_ub, max_y1_ub),
            xytext=(0.5, 0.9), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05, width=1),
            horizontalalignment='right', verticalalignment='top',
            )
axs1[0].legend()
axs1[1].legend()
plt.show()
```
